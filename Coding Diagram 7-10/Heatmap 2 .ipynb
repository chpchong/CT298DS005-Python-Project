```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Load the Excel file, assuming headers are in the second row (index 1)
data = pd.read_excel('21.xlsx', header=0)

# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Define relevant scoring columns
scoring_columns = ['Growth Scoring', 'Market Cap Scoring', 'ROA Scoring', 'PE Scoring', 'Total']

# Check if all scoring columns are present
missing_cols = [col for col in scoring_columns if col not in data.columns]
if missing_cols:
    print(f"Missing columns: {missing_cols}")
else:
    # Select the scoring data
    scoring_data = data[scoring_columns]

    # Sort the data by a specific column (e.g., 'Total')
    scoring_data = scoring_data.sort_values(by='Total', ascending=False)

    # Define a custom color palette
    colors = ['#39FF14', '#00FF00', '#32CD32', '#40E0D0', '#00FFFF', '#1B03A3']  # Neon green to neon blue
    custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

    # Create a heatmap with gridlines
    plt.figure(figsize=(10, 6))
    sns.heatmap(scoring_data, annot=True, cmap=custom_cmap, fmt='.1f', 
                yticklabels=data['Ticker'], 
                linecolor='black', linewidths=0.5)  # Add gridlines

    # Add titles and labels
    plt.title('Stock Pick Up in 2021')  # Updated title
    plt.xlabel('Scoring Type')
    plt.ylabel('Stock Ticker')

    # Show the plot
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Given simple returns for three investments
data = {
    'Year': [2022, 2023, 2024],
    'Stock Inv from 2022-2024 (%)': [3.33, 15.29, -29.33],
    '2800 (%)': [-8.35, -8.56, 28.85],
    'Risk Free Deposit (%)': [3.83, 4.44, 4.28]
}

df = pd.DataFrame(data)

# Plotting line chart for simple returns
plt.figure(figsize=(10, 6), facecolor='white')

# Plot lines for each investment's simple returns
for col in df.columns[1:]:
    plt.plot(df['Year'], df[col], marker='o', label=col)

# Customize the plot
plt.title('Annual Simple Returns Over Time', color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Simple Return (%)', color='black')
plt.xticks(color='black')
plt.yticks(color='black')
plt.grid(color='gray', linestyle='--', alpha=0.5)

# Show legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()

# Print simple returns for clarity
print("\nSimple Returns:")
print(df[['Year', 'Stock Inv from 2022-2024 (%)', '2800 (%)', 'Risk Free Deposit (%)']])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Given simple returns for three investments
data = {
    'Year': [2022, 2023, 2024],
    'Stock Inv from 2022-2024 (%)': [3.33, 15.29, -29.33],
    '2800 (%)': [-8.35, -8.56, 28.85],
    'Risk Free Deposit (%)': [3.83, 4.44, 4.28]
}

df = pd.DataFrame(data)

# Calculate log returns for each investment directly from simple returns
for col in df.columns[1:]:
    df['Log Returns ' + col] = np.log(1 + df[col] / 100)

# Calculate cumulative returns
cumulative_returns = {
    'Year': df['Year'],
    'Cumulative Return Stock Inv': [0] * len(df),
    'Cumulative Return 2800': [0] * len(df),
    'Cumulative Return Risk-Free Deposit': [0] * len(df)
}

# Initialize cumulative returns with the first year's log return
cumulative_returns['Cumulative Return Stock Inv'][0] = df['Log Returns Stock Inv from 2022-2024 (%)'][0]
cumulative_returns['Cumulative Return 2800'][0] = df['Log Returns 2800 (%)'][0]
cumulative_returns['Cumulative Return Risk-Free Deposit'][0] = df['Log Returns Risk Free Deposit (%)'][0]

# Calculate cumulative returns for subsequent years
for i in range(1, len(df)):
    cumulative_returns['Cumulative Return Stock Inv'][i] = cumulative_returns['Cumulative Return Stock Inv'][i-1] + df['Log Returns Stock Inv from 2022-2024 (%)'][i]
    cumulative_returns['Cumulative Return 2800'][i] = cumulative_returns['Cumulative Return 2800'][i-1] + df['Log Returns 2800 (%)'][i]
    cumulative_returns['Cumulative Return Risk-Free Deposit'][i] = cumulative_returns['Cumulative Return Risk-Free Deposit'][i-1] + df['Log Returns Risk Free Deposit (%)'][i]

# Create a DataFrame for cumulative returns
cumulative_df = pd.DataFrame(cumulative_returns)

# Combine log returns and cumulative returns
result_df = pd.concat([df[['Year', 'Log Returns Stock Inv from 2022-2024 (%)', 'Log Returns 2800 (%)', 'Log Returns Risk Free Deposit (%)']], cumulative_df], axis=1)

# Rename columns for clarity
result_df.columns = [
    'Year', 
    'Log Return Stock Investment (%)', 
    'Log Return 2800 (%)', 
    'Log Return Risk-Free Deposit (%)',
    'Year',  # Repeating Year column
    'Cumulative Return Stock Investment (%)', 
    'Cumulative Return 2800 (%)', 
    'Cumulative Return Risk-Free Deposit (%)'
]

# Format as percentages
result_df.iloc[:, 1:] = result_df.iloc[:, 1:] * 100

# Display the result
print(result_df)

# Plotting the bar chart for cumulative returns
plt.figure(figsize=(10, 6), facecolor='white')

# Set bar positions
x = np.arange(len(df))

# Bar width
bar_width = 0.25

# Define colors for consistency
colors = ['#39FF14', '#FFFF00', '#ADD8E6']  # Neon green, yellow, light blue

# Plot bars for cumulative returns with black edges
bars1 = plt.bar(x - bar_width, cumulative_df['Cumulative Return Stock Inv'] * 100, width=bar_width, label='Stock Investment', color=colors[0], edgecolor='black')
bars2 = plt.bar(x, cumulative_df['Cumulative Return 2800'] * 100, width=bar_width, label='2800 Investment', color=colors[1], edgecolor='black')
bars3 = plt.bar(x + bar_width, cumulative_df['Cumulative Return Risk-Free Deposit'] * 100, width=bar_width, label='Risk-Free Deposit', color=colors[2], edgecolor='black')

# Add cumulative return percentages above the bars in black
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{yval:.1f}%", ha='center', va='bottom', color='black')

# Add simple return percentages inside the bars in red
for i in range(len(df)):
    plt.text(bars1[i].get_x() + bars1[i].get_width()/2, bars1[i].get_height()/2, f"{df['Stock Inv from 2022-2024 (%)'][i]:.1f}%", ha='center', va='center', color='red')
    plt.text(bars2[i].get_x() + bars2[i].get_width()/2, bars2[i].get_height()/2, f"{df['2800 (%)'][i]:.1f}%", ha='center', va='center', color='red')
    plt.text(bars3[i].get_x() + bars3[i].get_width()/2, bars3[i].get_height()/2, f"{df['Risk Free Deposit (%)'][i]:.1f}%", ha='center', va='center', color='red')

# Customize the plot
plt.title('Cumulative Return from 2022-2024', color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Cumulative Return (%)', color='black')
plt.xticks(x, df['Year'], color='black')
plt.yticks(color='black')

# Set y-axis ticks to a shorter distance
plt.yticks(np.arange(-30, 35, 5))  # Adjust the range and interval as needed

plt.axhline(0, color='red', linewidth=1.5, linestyle='--')
plt.grid(color='gray', linestyle='--', alpha=0.5)

# Show legend
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout(pad=2.0)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Given simple returns for three investments
data = {
    'Year': [2022, 2023, 2024],
    'Stock Inv Return (%)': [3.33, 15.29, -29.33],
    'Stock 2800 (%)': [-8.35, -8.56, 28.85],
    'Risk-Free Deposit (%)': [3.83, 4.44, 4.28]
}

# Create DataFrame and ensure Year remains an integer
df = pd.DataFrame(data)
df['Year'] = df['Year'].astype(int)  # Keeps Year column as an integer

# Create and style the table
fig, ax = plt.subplots(figsize=(10, 6))  # Increased size for better readability
ax.set_axis_off()

# Define table colors
header_color = '#800080'  # Purple
cell_colors = ['#228B22', '#32CD32', '#90EE90']  # Lighter green for first row, transitioning lighter

# Create formatted table data with corrected headers including (%)
table_data = [
    ['Year', 'Stock Inv Return (%)', 'Stock 2800 (%)', 'Risk-Free Deposit (%)']
] + df.astype(str).values.tolist()  # Convert values to strings while keeping Year intact

# Create the table
table = ax.table(cellText=table_data, cellLoc='center', loc='center', colWidths=[0.2] * len(df.columns))

# Apply column header styles
for i in range(len(df.columns)):
    table[0, i].set_facecolor(header_color)
    table[0, i].set_text_props(color='white', weight='bold')

# Apply row color gradient with lighter first row
for i in range(1, len(df) + 1):
    for j in range(len(df.columns)):
        table[i, j].set_facecolor(cell_colors[i - 1])

# Adjust font size
table.auto_set_font_size(False)
table.set_fontsize(12)  # Increased font size

# Adjust column widths and scale for better readability
table.scale(1.2, 1.5)

plt.show()
```
