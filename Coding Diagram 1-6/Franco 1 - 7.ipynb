```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Define the stock indices and their Yahoo Finance tickers
indices = {
    'HSI': '^HSI',
    'NASDAQ': '^IXIC',
    'S&P 500': '^GSPC',
    'Nikkei 225': '^N225'
}

# Create an empty DataFrame to store results
df_all = pd.DataFrame()

# Fetch historical data for each index
for name, symbol in indices.items():
    print(f"Downloading data for {name} ({symbol})...")
    df = yf.download(symbol, start="2004-01-01", end="2024-04-01", interval="1mo")  # Monthly data
    
    # Check if data was retrieved successfully
    if not df.empty:
        df_resampled = df['Close'].resample('Y').last()  # Convert monthly data to yearly
        df_all[name] = df_resampled  # Store in main DataFrame

# Drop rows with missing values
df_all.dropna(inplace=True)

# Debugging: Check if data is successfully retrieved
print("\nData preview:\n", df_all.head())

# Define major financial crisis years
crisis_years = [2003, 2008, 2019]

# Plot historical price trends with yearly intervals
plt.figure(figsize=(12, 6))
for column in df_all.columns:
    plt.plot(df_all.index, df_all[column], marker='o', label=column)  # Add markers for clarity

# Add vertical lines for major financial crises
for year in crisis_years:
    plt.axvline(pd.Timestamp(str(year)), color='red', linestyle='--', linewidth=1, label=f'{year} Crisis' if year == crisis_years[0] else "")

plt.title('Historical Price Trends of Major Indices (2004-2024)')
plt.xlabel('Year')
plt.ylabel('Index Value')
plt.legend()
plt.grid()
plt.xticks(df_all.index, df_all.index.strftime('%Y'), rotation=45)  # Format x-axis to show years

plt.show()
```
