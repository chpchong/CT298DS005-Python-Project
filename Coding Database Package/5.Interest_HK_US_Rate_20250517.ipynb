# HK interest rate

```python
import numpy as np
import pandas as pd
import requests
import yfinance as yf

import matplotlib.pyplot as plt

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
apiFrom = '2020-01'
apiTo = '2024-12'

try:
    response = requests.get(f"https://api.hkma.gov.hk/public/market-data-and-statistics/monthly-statistical-bulletin/er-ir/hkd-ir-periodaverage?pagesize=1000&choose=end_of_month&from={apiFrom}&to={apiTo}&sortorder=desc")
    if response.status_code==200 : print('Connect successfully!')
    data = response.json()
except Exception as e:
    print(e)
```

```python
df_HKi = pd.DataFrame(data['result']['records'])
df_HKi = df_HKi.set_index('end_of_month')  
df_HKi.index = pd.to_datetime(df_HKi.index)
df_HKi = df_HKi.sort_values('end_of_month')
df_HKi.head(2)
```

```python
df_HKi.info()
```

```python
filtered_df = df_HKi.savings_deposit_rate[(df_HKi.index.month == 12)]
filtered_df
```

```python
filtered_df.to_csv('HK_deposit_rate.csv')  
```
