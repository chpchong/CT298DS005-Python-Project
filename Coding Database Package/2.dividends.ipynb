```python
import pandas as pd
import yfinance as yf
```

```python
tickers = pd.read_csv('stock_list.csv').Ticker
#tickers= ["1299.HK"]
tickers
```

```python
# Years we are interested in
years = [2021, 2022, 2023, 2024]

# Column names
columns = [f"Dividends_{year}" for year in years]

# Initialize an empty dict to store results
dividend_data = {}

for ticker in tickers:
    try:
        print(f"Fetching data for {ticker}...")
        stock = yf.Ticker(ticker)
        dividends = stock.dividends

        if not dividends.empty:
            # Convert index to datetime
            dividends.index = pd.to_datetime(dividends.index)
            annual_dividends = dividends.resample('YE').sum()
            annual_dividends.index = annual_dividends.index.year

            # Create entry for this ticker
            dividend_data[ticker] = {}
            for year in years:
                if year in annual_dividends.index:
                    dividend_data[ticker][f"Dividends_{year}"] = annual_dividends.loc[year]
                else:
                    dividend_data[ticker][f"Dividends_{year}"] = 0.0
        else:
            # No dividend data
            dividend_data[ticker] = {f"Dividends_{year}": 0.0 for year in years}
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        dividend_data[ticker] = {f"Dividends_{year}": 0.0 for year in years}

# Convert to DataFrame
df_dividends = pd.DataFrame.from_dict(dividend_data, orient='index')

```

```python
df_dividends.reset_index(inplace=True)
df_dividends.rename(columns={'index': 'Ticker'},inplace=True)
```
